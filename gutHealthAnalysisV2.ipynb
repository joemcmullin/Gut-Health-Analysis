```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# Load the dataset
file_path = 'metadata.tsv'  # Make sure this path is correct
metadata_df = pd.read_csv(file_path, sep='\t')

# Display the dataframe with updated styling
styled_df = metadata_df.style.set_table_styles([
    {'selector': 'th', 'props': [('background-color', '#f4f4f4'), ('color', 'black'), ('font-weight', 'bold')]},
    {'selector': 'td', 'props': [('text-align', 'center')]},
    {'selector': 'tr:nth-of-type(odd)', 'props': [('background-color', '#ddd')]},
    {'selector': 'tr:nth-of-type(even)', 'props': [('background-color', '#eee')]},
]).set_properties(**{'border': '1px solid black'}).hide(axis='index')

# Visualization: Line Chart of BMI Trends by Age for different Surgery Types
plt.figure(figsize=(12, 8))
sns.lineplot(data=metadata_df, x='Age', y='BMI', hue='Surgery_Type', marker='o')
plt.title('BMI Trend by Age and Surgery Type')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.legend(title='Surgery Type')
plt.grid(True)
plt.show()

# Count of Surgery Types by Gender
plt.figure(figsize=(10, 6))
sns.countplot(data=metadata_df, x='Surgery_Type', hue='Gender')
plt.title('Count of Surgery Types by Gender')
plt.xlabel('Surgery Type')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.xticks(rotation=45)
plt.show()

# Check column names
print(metadata_df.columns)

# Assuming 'Surgery_Type' and 'Drinking Status' are correct column names
# Update these variable names if necessary based on the actual column names in your DataFrame
surgery_type_col = 'Surgery_Type'
Drinking Status_col = 'Drinking Status'

# Creating an interactive bar chart with Plotly for Surgery Types by Drinking Status and Gender
# Adjust column names if necessary
counts_df = metadata_df.groupby([surgery_type_col, Drinking Status_col, 'Gender']).size().reset_index(name='Counts')

fig = px.bar(counts_df, x=surgery_type_col, y='Counts', color=Drinking Status_col,
             barmode='group', facet_row='Gender',
             title='Counts of Surgery Types by Drinking Status and Gender',
             labels={'Counts': 'Number of Cases', surgery_type_col: 'Surgery Type', Drinking Status_col: 'Drinking Status'},
             height=600)
fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.update_yaxes(matches=None)
fig.show()

# display top 10 rows of data
display(metadata_df.head(10))
```

```python
import pandas as pd
import plotly.express as px

# Load the dataset
file_path = 'metadata.tsv'  # Ensure this path is correct
metadata_df = pd.read_csv(file_path, sep='\t')

# Interactive Line Chart: BMI Trends by Age for different Surgery Types using Plotly
# Note: Here you can modify the 'color' argument for different themes
fig_line = px.line(metadata_df, x='BMI', y='Surgery_Type', color='Age', markers=True,
                   title='BMI Trend by Age and Surgery Type',
                   labels={'BMI': 'BMI', 'Age': 'Age', 'Surgery_Type': 'Surgery Type'})
fig_line.update_layout(xaxis_title='Age', yaxis_title='BMI', legend_title='Surgery Type')
# To change the color theme, you can use color_continuous_scale or color_discrete_sequence parameters.
fig_line.show()

# Preparing data for the interactive count plot by counting occurrences
counts_surgery_gender = metadata_df.groupby(['Surgery_Type', 'Gender']).size().reset_index(name='Counts')

# Interactive Bar Chart: Count of Surgery Types by Gender using Plotly
# Note: Modify 'color' in px.bar() for different color themes
fig_count = px.bar(counts_surgery_gender, x='Surgery_Type', y='Counts', color='Gender', barmode='group',
                   title='Count of Surgery Types by Gender',
                   labels={'Counts': 'Number of Cases', 'Surgery_Type': 'Surgery Type', 'Gender': 'Gender'})
fig_count.update_layout(xaxis_title='Surgery Type', yaxis_title='Count', legend_title='Gender', xaxis={'categoryorder':'total descending'})
# To change the color theme, you can use color_continuous_scale or color_discrete_sequence parameters.
fig_count.show()

# Interactive Bar Chart for Surgery Types by Drinking Status and Gender
# Assuming 'Drinking Status' is a correct column name, adjust as necessary
counts_df = metadata_df.groupby(['Surgery_Type', 'Drinking Status', 'Gender']).size().reset_index(name='Counts')

# Note: Modify 'color' in px.bar() for different color themes
fig_interactive = px.bar(counts_df, x='Surgery_Type', y='Counts', color='Drinking Status',
                         barmode='group', facet_row='Gender',
                         title='Counts of Surgery Types by Drinking Status and Gender',
                         labels={'Counts': 'Number of Cases', 'Surgery_Type': 'Surgery Type', 'Drinking Status': 'Drinking Status'},
                         height=600)
fig_interactive.update_layout(xaxis_title='Surgery Type', yaxis_title='Count', legend_title='Drinking Status', xaxis={'categoryorder':'total descending'})
# To customize themes and colors, consider exploring Plotly's templates with fig_interactive.layout.template = '...'
# And color schemes with fig_interactive.update_traces(marker=dict(color=...)) for specific color adjustments
fig_interactive.show()

# Display top 10 rows of the DataFrame
# This display uses pandas and is not inherently interactive
display(metadata_df.head(10))
```
